To forecast the net income and profitability of the startup in the following 12 months, we need to consider the following factors: 

*Monthly revenue per customer:* 

This is calculated based on the interest rate and the loan amount for each Contigo product. 


- Contigo A: $1000*0.15=150$ 


- Contigo B: $2000*0.10=200$


*Net income per customer:* 

This is the revenue per customer minus the operative cost per account

- Contigo A: $150-30=120$


- Contigo B: $200-30=170$


*Debt delinquency (DQ) rate:* 

This refers to the percentage of loans for each credit portfolio whose payments are delinquent. The forecast takes into consideration the accounts that went delinquent for each product. Contigo A 13.5% and 8% for Contigo B. 

Customers in the month 1 Contigo A= 5000; Contigo B= 10,000

- Contigo A

Loans whose payments are delinquent:  $5000*0.135= 675$

Loans whose payments are not delinquent: $5000*(1-0.135)= 4325$


- Contigo B

Loans whose payments are delinquent:  $10000*0.08= 800$

Loans whose payments are not delinquent: $10000*(1-0.08)=9200$ 


Using these calculations, we can forecast the net income and profitability of the startup for the next 12 months. 

```python
# Net Income for the next 12 months 
customer_CA=5000
customer_CB=10000
DQ_CA=0.135
DQ_CB=0.08
int_rateA=0.15
int_rateB=0.10
grow_rate_CA=0.25
grow_rate_CB=0.10
loan_A=1000
loan_B=2000

net_income=[]
month=[]
for m in range(1,13):
    
    month.append(m)
    # Expected revenue
    exp_rev_CA=customer_CA*(1-DQ_CA)*(int_rateA*loan_A)
    exp_rev_CB=customer_CB*(1-DQ_CB)*(int_rateB*loan_B)
    total_rev=exp_rev_CA+exp_rev_CB

    # Expected losses 
    exp_loss_CA=customer_CA*(DQ_CA)*1000
    exp_loss_CB=customer_CB*(DQ_CB)*2000
    total_loss=exp_loss_CA+exp_loss_CB


    # Operative cost 
    OC=(customer_CA+customer_CB)*30

    # Net income
    net_income.append(total_rev-total_loss-OC)
    
    customer_CA*=(1+grow_rate_CA)
    customer_CB*=(1+grow_rate_CB)

for i,j in zip(month,net_income):
    print(f"Month {i}: Net Expected Income= {j:.0f}")
```

After calculating the total expected income we can see this startup is not going to be profitable during the next months in order to understand the trends in the Contigo products we forecast the expected revenue and losses per month and product.  

```python
import pandas as pd

# Create months 
month=[]
for m in range(1,13):
    month.append(m)
    
# Create contigo A customers over 12 months - grow rate of 25% - 
CA_initial=5000
grow_rate_CA=0.25
accounts_CA=[]
for i in range(12):
    accounts_CA.append(round(CA_initial,0))
    CA_initial+=CA_initial*grow_rate_CA
    
# Create contigo B customers over 12 months - grow rate of 10% - 
CB_initial=10000
grow_rate_CB=0.10
accounts_CB=[]
for i in range(12):
    accounts_CB.append(round(CB_initial,0))
    CB_initial+=CB_initial*grow_rate_CB
    
# Create a dictionary with monts and Contigo A y B customers
contigo_dict={'month':month,
             'accounts_CA':accounts_CA,
              'accounts_CB':accounts_CB
             }

# Create a DataFrame
df=pd.DataFrame(contigo_dict)
df.set_index('month',inplace=True)
```

```python
# Create variables of interest
#DQ accounts 
df['DQ_accountsA']=df['accounts_CA']*0.135
df['DQ_accountsB']=df['accounts_CB']*0.08
# Loan Contigo A y B customers
df['Loan_A']=df['accounts_CA']*1000
df['Loan_B']=df['accounts_CB']*2000
# Expected revenue Contigo A y B 
df['Expected_revenue_A']=(df['accounts_CA']*(1-0.135))*(0.15*1000)
df['Expected_revenue_B']=(df['accounts_CB']*(1-0.08))*(0.10*2000)
# Expected loss Contigo A y B  
df['Expected_loss_A']=(df['accounts_CA']*(0.135))*1000
df['Expected_loss_B']=(df['accounts_CB']*(0.08))*2000
# Total Expected loss
df['Expected_loss']=df['Expected_loss_A']+df['Expected_loss_B']
# Operative monthly cost Contigo A 
df['Operative_cost_A']=df['accounts_CA']*30
# Operative monthly cost Contigo B
df['Operative_cost_B']=df['accounts_CB']*30
# Total operative monthly cost 
df['Operative_cost']=(df['accounts_CA']+df['accounts_CB'])*30
# Net income Contigo A
df['Net_income_A']=df['Expected_revenue_A']-df['Operative_cost_A']-df['Expected_loss_A']
# Net income Contigo B
df['Net_income_B']=df['Expected_revenue_B']-df['Operative_cost_B']-df['Expected_loss_B']
# Total Net Income
df['Net_income']=df['Expected_revenue_A']+df['Expected_revenue_B']-df['Operative_cost']-df['Expected_loss']
```

Since the company is not profitable in Contigo A y B products with the current interest rates the startup can adjust the interest rates.

```python
def net_income(int_rateA,int_rateB):

    customer_CA=5000
    customer_CB=10000
    DQ_CA=0.135
    DQ_CB=0.08
    grow_rate_CA=0.25
    grow_rate_CB=0.10
    loan_A=1000
    loan_B=2000

    net_income=[]
    month=[]
    for m in range(1,13):

        month.append(m)
        # Expected revenue
        exp_rev_CA=(customer_CA*(1-DQ_CA))*(int_rateA*loan_A)
        exp_rev_CB=(customer_CB*(1-DQ_CB))*(int_rateB*loan_B)
        total_rev=exp_rev_CA+exp_rev_CB

        # Expected losses 
        exp_loss_CA=customer_CA*(DQ_CA)*1000
        exp_loss_CB=customer_CB*(DQ_CB)*2000
        total_loss=exp_loss_CA+exp_loss_CB


        # Operative cost 
        OC=(customer_CA+customer_CB)*30

        # Net income
        net_income.append(total_rev-total_loss-OC)

        customer_CA*=(1+grow_rate_CA)
        customer_CB*=(1+grow_rate_CB)

    for i,j in zip(month,net_income):
        print(f"Month {i}: Net Expected Income= {j:.0f}")
```
